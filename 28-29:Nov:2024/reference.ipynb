```python

import pandas as pd

# Function to load data from a CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

# Function to select ATM CE and PE strikes based on timestamps
def select_atm_strikes(data, Time):
    atm_strikes = {}
    for Time in Time:
        row = data[data['Time'] == Time]
        
        if not row.empty:
            atm_ce = row['ATM_CE'].values[0]  # Replace 'ATM_CE' with your column name
            atm_pe = row['ATM_PE'].values[0]  # Replace 'ATM_PE' with your column name
            atm_strikes[Time] = (atm_ce, atm_pe)
    return atm_strikes

# Function to execute trades and calculate profit/loss
def execute_trades(data, atm_strikes):
    results = []
    
    for Time, (atm_ce, atm_pe) in atm_strikes.items():
        entry_ce = data.loc[data['Time'] == Time, 'Close_CE'].values[0]  # Close price for CE
        entry_pe = data.loc[data['Time'] == Time, 'Close_PE'].values[0]  # Close price for PE
        
        stoploss_ce = entry_ce * 2  # 100% stoploss
        stoploss_pe = entry_pe * 2  # 100% stoploss
        
        # Simulate trading
        for index, row in data.iterrows():
            if row['Time'] > Time and row['Time'] <= '23:26:59':
                print(f'Current Time: {row["Time"]}, Close_CE: {row["Close_CE"]}, Close_PE: {row["Close_PE"]}')  # Debugging line
                if row['Close_CE'] >= stoploss_ce:
                    exit_pe = row['Close_PE']
                    pl_pe = entry_pe - exit_pe
                    print(f'Stoploss hit for PE: {pl_pe}')  # Debugging line
                    results.append((Time, 'PE', pl_pe))
                    break
                elif row['Close_PE'] >= stoploss_pe:
                    exit_ce = row['Close_CE']
                    pl_ce = entry_ce - exit_ce
                    print(f'Stoploss hit for CE: {pl_ce}')  # Debugging line
                    results.append((Time, 'CE', pl_ce))
                    break
            else:
                print('No stoploss hit, calculating final P&L...')  # Debugging line
                final_ce = data.loc[data['Time'] == '23:26:59', 'Close_CE'].values[0]
                final_pe = data.loc[data['Time'] == '23:26:59', 'Close_PE'].values[0]
                pl_ce = entry_ce - final_ce
                pl_pe = entry_pe - final_pe
                print(f'Final P&L for CE: {pl_ce}, PE: {pl_pe}')  # Debugging line
                results.append((Time, 'CE', pl_ce))
                results.append((Time, 'PE', pl_pe))

    return results

# Function to save results back to CSV
def save_results(file_path, results):
    results_df = pd.DataFrame(results, columns=['Timestamp', 'Leg', 'P&L'])
    results_df.to_csv(file_path, index=False)

# Main function
def main():
    file_path = 'Jan_data_2022.csv'  # Your CSV file path
    data = load_data(file_path)
    
#     Input timestamps (for example)
    Time = [
         "09:39:59",
    "09:59:59",
    "10:29:59",
    "10:59:59",
    "11:29:59",
    "11:59:59",
    "12:29:59",
    "12:59:59",
    "13:29:59",
    "13:59:59"
    ]
    
    atm_strikes = select_atm_strikes(data, Time)
    results = execute_trades(data, atm_strikes)
    
    save_results('trading_results1.csv', results)

if __name__ == "__main__":
    main()
```
