```python
# Loading data needed to test our theory
ce_data = data['2024-01-01']['2024-02-01'][23000]['CE']
print(ce_data)
```

```python
# using necessary features we need
close_prices = ce_data[['time', 'close']]
print(close_prices)
```

```python
import pandas as pd


close_prices['time'] = pd.to_datetime(close_prices['time'])

# Setting the entry time and universal exit time
entry_time = pd.to_datetime('09:16:00').time()
universal_exit_time = pd.to_datetime('15:28:00').time()

# Filtering the entry data
entry_data = close_prices[close_prices['time'].dt.time == entry_time]

if not entry_data.empty:
    entry_close_price = entry_data.iloc[0]['close']
    entry_time = entry_data.iloc[0]['time'].time()  
    print(f"Entry Time: {entry_time}, Entry Close Price: {entry_close_price}")

    # Define the half price for exit condition
    half_price_target = entry_close_price / 2

    # Filter the exit data based on conditions:
    # Exit when the price is half of the entry close price or at the universal exit time
    exit_data = close_prices[(close_prices['time'].dt.time > entry_time) & 
                             ((close_prices['close'] <= half_price_target) | 
                              (close_prices['time'].dt.time == universal_exit_time))]

    if not exit_data.empty:
        exit_data = exit_data.sort_values(by='time') 
        exit_close_price = exit_data.iloc[0]['close']
        exit_time = exit_data.iloc[0]['time'].time()

         # Display the exit time, close price, and calculate the profit/loss from the trade
        print(f"Exit Time: {exit_time}, Exit Close Price: {exit_close_price}")
        print(f"Profit/Loss from Entry to Exit: {entry_close_price - exit_close_price}")

```

```python
import pandas as pd

# Assuming 'data' is a nested dictionary structured as data[date][expiry][strike][CE/PE]
# We would replace this with the actual data loading or initialization

# Displaying available dates and take input
print("Available Dates:", list(data.keys()))
date = input("Enter the date (e.g., '2024-01-01'): ")

# Displaying available expiries for the selected date and take input
print("Available Expiries:", list(data[date].keys()))
expiry = input("Enter the expiry date (e.g., '2024-02-01'): ")

# Displaying available strike prices for the selected date and expiry and take input
print("Available Strike Prices:", list(data[date][expiry].keys()))
strike = int(input("Enter the strike price (e.g., 23000): "))

# Displaying options for CE or PE and take input
print("Available Options: CE, PE")
option_type = input("Enter option type (CE/PE): ")

# Accessing the selected data
selected_data = data[date][expiry][strike][option_type]

# Assuming 'selected_data' is a DataFrame with the OHLCV columns
# Converting the 'time' column to datetime format
selected_data['time'] = pd.to_datetime(selected_data['time'])

# Setting the entry time and universal exit time
entry_time = pd.to_datetime(input("Enter entry time (HH:MM:SS, e.g., '09:16:00'): ")).time()
universal_exit_time = pd.to_datetime(input("Enter universal exit time (HH:MM:SS, e.g., '15:28:00'): ")).time()

# Filtering for entry data
entry_data = selected_data[selected_data['time'].dt.time == entry_time]

if not entry_data.empty:
    entry_close_price = entry_data.iloc[0]['close']
    print(f"Entry Time: {entry_data.iloc[0]['time'].time()}, Entry Close Price: {entry_close_price}")

    # Defining the half price for exit condition
    half_price_target = entry_close_price / 2

    # Filtering the exit data based on conditions:
    exit_data = selected_data[(selected_data['time'].dt.time > entry_time) & 
                              ((selected_data['close'] <= half_price_target) | 
                               (selected_data['time'].dt.time == universal_exit_time))]

    if not exit_data.empty:
        exit_data = exit_data.sort_values(by='time') 
        exit_close_price = exit_data.iloc[0]['close']
        exit_time = exit_data.iloc[0]['time'].time()

        # Displaying the exit time, close price, and calculate the profit/loss from the trade
        print(f"Exit Time: {exit_time}, Exit Close Price: {exit_close_price}")
        print(f"Profit/Loss from Entry to Exit: {entry_close_price - exit_close_price}")
    else:
        print("No valid exit condition met.")
else:
    print("No entry data available at the specified time.")
```
