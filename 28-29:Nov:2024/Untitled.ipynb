```python
import pandas as pd
import datetime

# Function to load data from a CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

# Function to get close price for the option based on ticker and timestamp
def get_close_price(data, ticker, timestamp):
    row = data[(data['Ticker'] == ticker) & (data['Time'] == timestamp)]
    if not row.empty:
        return row['Close'].values[0]
    else:
        print(f"Missing data for {ticker} at {timestamp}")
        return None

# Function to execute trades and calculate profit/loss
def execute_trades(data, initial_date, initial_time):
    # Convert initial_time to a time object
    if isinstance(initial_time, str):
        initial_time = datetime.datetime.strptime(initial_time, "%H:%M:%S").time()

    # Filtering data to get relevant entries
    filtered_entries = data[(data['Date'] == initial_date) & (data['Ticker'] == 'CRUDEOIL-I.MCX')]
    filtered_entries_after_time = filtered_entries[filtered_entries['Time'] >= str(initial_time)]
    
    if filtered_entries_after_time.empty:
        print(f"No data available after {initial_time} on {initial_date}")
        return []

    nearest_time_entry = filtered_entries_after_time.iloc[0]
    spot_price = int(nearest_time_entry['Spot'])

    # Constructing the tickers for the CE and PE options at the ATM strike price
    options_ce_ticker = f'CRUDEOIL23APR{spot_price}CE.MCX'
    options_pe_ticker = f'CRUDEOIL23APR{spot_price}PE.MCX'

    # Getting initial close prices for both CE and PE options
    entry_ce = get_close_price(data, options_ce_ticker, nearest_time_entry['Time'])
    entry_pe = get_close_price(data, options_pe_ticker, nearest_time_entry['Time'])

    if entry_ce is None or entry_pe is None:
        print(f"Skipping timestamp {nearest_time_entry['Time']} due to missing close price data.")
        return []

    # Setting stop loss for both options (100% stop loss for initial entry)
    stoploss_ce = entry_ce * 2
    stoploss_pe = entry_pe * 2

    # Simulating the trade loop
    results = []
    current_ce_position = entry_ce
    current_pe_position = entry_pe
    last_entry_time = nearest_time_entry['Time']
    position_1 = 'CE'  # Initially enter CE
    position_2 = 'PE'  # Initially enter PE

    # Time to stop (universal exit time)
    t_s = '23:26:59'  # Universal exit time

    # Loop through the data to simulate trading
    for index, row in data.iterrows():
        # Skip rows before or at the universal exit time
        if row['Time'] > last_entry_time and row['Time'] <= t_s:
            print(f"Current Time: {row['Time']}, Close_CE: {row['Close']}, Close_PE: {row['Close']}")
            
            # Track PE and CE positions and check if stop loss is hit
            if position_1 == 'PE' and row['Close'] >= stoploss_pe:
                # Exit PE and take a new CE position
                exit_pe = row['Close']
                pl_pe = entry_pe - exit_pe
                print(f"Stoploss hit for PE: {pl_pe}")
                results.append(('PE', pl_pe, row['Time']))
                # Enter new CE position, but keep PE position
                position_1 = 'CE'
                current_ce_position = get_close_price(data, f'CRUDEOIL23APR{spot_price}CE.MCX', row['Time'])
                stoploss_ce = current_ce_position * 2  # 100% stop loss
                entry_ce = current_ce_position
                last_entry_time = row['Time']

            elif position_1 == 'CE' and row['Close'] >= stoploss_ce:
                # Exit CE and take a new PE position
                exit_ce = row['Close']
                pl_ce = entry_ce - exit_ce
                print(f"Stoploss hit for CE: {pl_ce}")
                results.append(('CE', pl_ce, row['Time']))
                # Enter new PE position, but keep CE position
                position_1 = 'PE'
                current_pe_position = get_close_price(data, f'CRUDEOIL23APR{spot_price}PE.MCX', row['Time'])
                stoploss_pe = current_pe_position * 2  # 100% stop loss
                entry_pe = current_pe_position
                last_entry_time = row['Time']

            # Track the third option (the second option after the stop loss) and check for its stop loss
            if position_2 == 'PE' and row['Close'] >= stoploss_pe:
                exit_pe = row['Close']
                pl_pe = entry_pe - exit_pe
                print(f"Stoploss hit for second PE: {pl_pe}")
                results.append(('PE', pl_pe, row['Time']))
                break
            elif position_2 == 'CE' and row['Close'] >= stoploss_ce:
                exit_ce = row['Close']
                pl_ce = entry_ce - exit_ce
                print(f"Stoploss hit for second CE: {pl_ce}")
                results.append(('CE', pl_ce, row['Time']))
                break

    # If no stop loss is triggered by the universal exit time, calculate final P&L
    if row['Time'] == t_s:
        final_ce = get_close_price(data, f'CRUDEOIL23APR{spot_price}CE.MCX', t_s)
        final_pe = get_close_price(data, f'CRUDEOIL23APR{spot_price}PE.MCX', t_s)
        pl_ce = entry_ce - final_ce
        pl_pe = entry_pe - final_pe
        print(f"Final P&L at {t_s} - CE: {pl_ce}, PE: {pl_pe}")
        results.append(('Final', pl_ce + pl_pe, t_s))

    return results

# Function to save results back to CSV
def save_results(file_path, results):
    results_df = pd.DataFrame(results, columns=['Option', 'PnL', 'Time'])
    results_df.to_csv(file_path, index=False)

# Main function
def main():
    file_path = 'Filtered_CRUDEOIL_Data.csv'  # Your CSV file path
    data = load_data(file_path)
    
    initial_date = '03/04/2023'
    initial_time = '09:20:59'  # Defined time for the timestamp filtering
    
    results = execute_trades(data, initial_date, initial_time)
    
    save_results('trading_results.csv', results)

if __name__ == "__main__":
    main()
```
